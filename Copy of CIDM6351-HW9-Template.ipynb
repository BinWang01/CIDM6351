## **Q1 (1 points)**: load the graph from ```Homework9.edgelist``` file as graph ```G``` and print summary information of this graph ```G```

````python
G = nx.read_edgelist('Homework9.edgelist',  data=False, nodetype= int)
# print network summary information
print(G)




````

## **Q2 (2 points)**: Plot this graph ```G``` (1 point)

````python
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)




````

````python
# sample output:
````

## **Q3 (3 point)**: plot the network based on betweenness values the more important a node is, the larger node size it has


````python
G_betweenness = nx.betweenness_centrality(G)
nx.draw(G, pos, 
        nodelist= list(G_betweenness.keys()), 
        node_size=[v * 5000+10 for v in list(G_betweenness.values())], 
        with_labels=True,
        font_size=16
       )




````

````python
# sample output
````

## **Q4 (4 points)**: Find Communities in a Network and plot them in different colors

````python
communities = nx.algorithms.community.greedy_modularity_communities(G)
betweenness = nx.betweenness_centrality(G)
colors = ['deepskyblue','orange','limegreen','deeppink','yellow']
shapes = ['s','o','^','v','<']
for community,color, shape in zip(communities,colors, shapes):
    nx.draw_networkx_nodes(G,pos,nodelist=community,node_color=color, node_shape=shape, node_size=[betweenness[s]*3000+100 for s in community])
nx.draw_networkx_labels(G,pos)
nx.draw_networkx_edges(G, pos,edge_color = 'gray')




````
