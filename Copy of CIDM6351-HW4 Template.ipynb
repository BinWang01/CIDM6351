### Q1 (5 points) Get stock info and print out following information of each stock:
- full name
- sector 
- business summary
- website

````
# run this code cell to install required packages at first
# install yfinace
!pip install yfinance 
# update plotly to current version and we will use it for data visualization
!pip install plotly --upgrade
#  import yfinace for retrieving data
import yfinance as yf
# import plotly for ploting data
import plotly.express as px
````

````
# ------Put your answer below this line-----
apple = yf.Ticker("AAPL")
apple_info = apple.info
# convert msft_info to Python dictionary data structre and print it out with pprint
print(f"Full Name: {apple_info['longName']}\n")
print(f"Stock Sector: {apple_info['sector']}\n")
print(f"Business Summary: {apple_info['longBusinessSummary']}\n")
print(f"Website: {apple_info['website']}\n")
print("---------------------------")

netflix = yf.Ticker("NFLX")
netflix_info = netflix.info
# convert msft_info to Python dictionary data structre and print it out with pprint
print(f"Full Name: {netflix_info['longName']}\n")
print(f"Stock Sector: {netflix_info['sector']}\n")
print(f"Business Summary: {netflix_info['longBusinessSummary']}\n")
print(f"Website: {netflix_info['website']}\n")
print("---------------------------")

msft = yf.Ticker("MSFT")
msft_info = msft.info
# convert msft_info to Python dictionary data structre and print it out with pprint
print(f"Full Name: {msft_info['longName']}\n")
print(f"Stock Sector: {msft_info['sector']}\n")
print(f"Business Summary: {msft_info['longBusinessSummary']}\n")
print(f"Website: {msft_info['website']}\n")
print("---------------------------")

twitter = yf.Ticker("TWTR")
twitter_info = twitter.info
# convert msft_info to Python dictionary data structre and print it out with pprint
print(f"Full Name: {twitter_info['longName']}\n")
print(f"Stock Sector: {twitter_info['sector']}\n")
print(f"Business Summary: {twitter_info['longBusinessSummary']}\n")
print(f"Website: {twitter_info['website']}\n")
print("---------------------------")
````

## Q2 (5 points) Get history stock price data ```start="2022-01-01", end="2023-01-01" ``` for each stock and plot the line chart for the ```High``` stock price
Hint: https://aroussi.com/post/python-yahoo-finance

````
# ------Put your answer below this line-----

history_data = yf.Tickers("APPL NFLX MSFT TWTR").history(start="2022-01-01", end="2023-01-01")['High']
fig = px.line(history_data,title='High Stock Pirce 2022-01-01 to 2023-01-01',
              labels={"value":"Stock Price ($)", "variable":"Price Type"})
fig.show()

````
