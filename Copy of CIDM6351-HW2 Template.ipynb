## Q2 (2 points): Read the **HW2.json** in Python and print out the **name** and the **Audience score** for movies that produced **after 2010**. (Following is the sample content of the JSON file).
```json
{
    "Audience score %": "72",
    "Film": "Water For Elephants",
    "Genre": "Drama",
```

````
# ------Put your answer below this line-----
import json
with open('HW2.json') as f:
  data = json.load(f)
for element in data:
  if int(element['Year'])>2010:
    print(f"Name: {element['Film']} \nAudience score %: {element['Audience score %']}\n-----------")






````
