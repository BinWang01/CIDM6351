## Run this cell to load data from csv files

````
import pandas as pd
df_stu1 = pd.read_csv('https://raw.githubusercontent.com/czwtamu/CIDM6351/main/stu_info1.csv')
df_stu2 = pd.read_csv('https://raw.githubusercontent.com/czwtamu/CIDM6351/main/stu_info2.csv')
df_course = pd.read_csv('https://raw.githubusercontent.com/czwtamu/CIDM6351/main/course_info.csv')
df_grade = pd.read_csv('https://raw.githubusercontent.com/czwtamu/CIDM6351/main/course_grade.csv')
````

## Q1(1 point) Concat dataframes ```df_stu1``` and ```df_stu2```, and save the result to ```df_stu```. Show first 5 rows

````
# ------Put your answer below this line-----
df_stu = pd.concat([df_stu1,df_stu2],axis=0)
df_stu.head()


````

## Q2 (2 points) 
- 1 point: merge (inner) the student information, course information and students' grades
- 1 point: drop the coulumn of course id
- rename the coulumn names (optional)
- show first 5 rows

````
# ------Put your answer below this line-----
df_stu_grade = df_stu.merge(df_grade, on = 'stu_id', how='inner')
df_stu_grade_course = df_stu_grade.merge(df_course, on ='course_id', how='inner')
df_stu_grade_course.drop('course_id', axis=1,inplace=True)
df_stu_grade_course2 = df_stu_grade_course.rename({'stu_id':'StuID','stu_name':'StuName','address':'City','birthday':'Birthday','course_grade':'Grade','course_name':'CourseName','course_format':'Format'}, axis=1)
df_stu_grade_course2.head(5)

````

## Q4 (4 points) 
- 1 point: select students born between ```1/1/2008``` and ```12/31/2009``` ( you may need to update the datatype of Birthday to datetime)
- 2 points: **then** find the ```max``` and ```min``` grades of ```online``` courses for each selected student. 
- 1 point: Finally, rename the max and min garde to ```'OnlineMax','OnlineMin'``` and print out

````
# ------Put your answer below this line-----
df_stu_grade_course2 =  df_stu_grade_course2.astype({'Birthday': 'datetime64'})
condition1 = (df_stu_grade_course2['Birthday'] > '2008-01-01') & (df_stu_grade_course2['Birthday'] <= '2009-12-31')
df_subset1 = df_stu_grade_course2.loc[condition1]
condition2 = df_subset1['Format'] == 'online'
df_subset2 = df_subset1.loc[condition2]
df_subset2.groupby(
   ['StuName']
).agg(
    OnlineMax = ('Grade', max),
    OnlineMin = ('Grade', min)
)
````
