## Q4 (2 point) using ```.loc()``` method to filter all movies produced after ```2007``` and slice a sub-dataset that contains following 4 columns: 
```Lead Studio,	Audience score %,	Profitability, Year```

````
# ------Put your answer below this line-----
condition = df['Year'] > 2007
column_names = ['Lead Studio','Audience score %','Profitability','Year']
df1 = df.loc[condition,column_names]
df1
````

````
# sample output
````

## Q5 (2 points) Based on last step, show the ```max``` Audience score for each Studio on each year

````
# ------Put your answer below this line-----
df1.groupby(['Lead Studio','Year'])['Audience score %'].max()
````

### Q6 (3 point)  Groupby with ```sudio name``` and ```year```, then calculate:
- mean for Audience score
- min and max for Profitability

Rename the result to AudiMean, ProfMax, ProfMin

````
# ------Put your answer below this line-----
agg_result = df1.groupby(['Lead Studio','Year']).agg(
    {
        'Audience score %':'mean',
        'Profitability':['max','min']
    }
)
agg_result
````
